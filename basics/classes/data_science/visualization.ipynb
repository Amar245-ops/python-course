```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
plt.style.use('seaborn')
```

```python
data = np.random.randint(0, 100, 50)
print(data)
```

```python
plt.figure(figsize=(15, 5))
plt.plot(data, 
color='red', 
linewidth=3, 
linestyle=':', 
marker='*',
markersize = 10)
```

```python
plt.figure(figsize=(25, 5), facecolor='white')
plt.plot(data, 'g--x')
plt.title("Simple line plot demo", fontsize=20, color='red')
plt.xlabel(" X axis wala data")
plt.ylabel(" Y axis wala data")
```

```python
fig = plt.figure(figsize=(25, 5))
plt.plot(data, 'g--x')
names = ['A','B','C','D','E','F','G','H','I','J','K']
plt.xticks(np.arange(0, 51, 5), labels=names, rotation=45, color='red',fontsize=16)
plt.show()
```

```python
data1 = np.random.randint(0,100,50)
data2 = np.random.randint(0,100,50)
fig = plt.figure(figsize=(25,5))
plt.subplot(2,1,1)
plt.plot(data1,'r:',label='data1')
plt.subplot(2,1,2)
plt.plot(data2,label='data2')
plt.legend(loc='lower right',fontsize=16)
plt.show()
```

```python
fix,ax = plt.subplots(figsize=(25,10),nrows=2,ncols=1, facecolor='white', sharex=True)
ax[0].plot(data1,'rx:',label='data1',)
ax[1].plot(data2,'bo--',label='data2',)
ax[0].legend(loc='lower right',fontsize=16)
ax[1].legend(loc='lower right',fontsize=16)
ax[0].grid(True)
ax[1].grid(True)
plt.show()
```
